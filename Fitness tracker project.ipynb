```python
df["Age"] = df["Age"].astype(int)
```

```python
df["Gender"]=df["Gender"].astype(str)
```

```python
df["Gender"]=df["Gender"].str.strip().str.capitalize()
```

```python
df["Gender"].fillna("Unknown",inplace=True)
```

```python
print( df["Gender"].isnull().sum())
```

```python
df["Steps"] = df["Steps"].astype(float)
```

```python
df["Steps"].fillna(df["Steps"].median(), inplace=True)
```

```python
df.loc[(df["Steps"] < 0),"Steps"]= df["Steps"].median()
```

```python
df.loc[(df["Steps"] > 1000000),"Steps"]= df["Steps"].median()
```

```python
df["Steps"] = df["Steps"].astype(int)
```

```python
print(df["Steps"].describe())
```

```python
print( df["Steps"].isnull().sum())
```

```python
df["Calories_Burned"] = df["Calories_Burned"].astype(float)
```

```python
df["Calories_Burned"].fillna(df["Calories_Burned"].median(), inplace=True)
```

```python
df.loc[(df["Calories_Burned"] < 0) , "Calories_Burned"]=df["Calories_Burned"].median()
```

```python
df.loc[(df["Calories_Burned"] > 10000) , "Calories_Burned"]=df["Calories_Burned"].median()
```

```python
df["Calories_Burned"] = df["Calories_Burned"].astype(int)
```

```python
print(df["Calories_Burned"].describe())
```

```python
print( df["Calories_Burned"].isnull().sum())
```

```python
df["Workout_Duration_Min"]=df["Workout_Duration_Min"].astype(float)
```

```python
df["Workout_Duration_Min"].fillna(df["Workout_Duration_Min"].median(), inplace=True)
```

```python
df.loc[(df["Workout_Duration_Min"] < 0), "Workout_Duration_Min"]= df["Workout_Duration_Min"].median()
```

```python
df.loc[(df["Workout_Duration_Min"] > 1440), "Workout_Duration_Min"]= df["Workout_Duration_Min"].median()
```

```python
df["Workout_Duration_Min"] = df["Workout_Duration_Min"].astype(int)
```

```python
print(df["Workout_Duration_Min"].describe())
```

```python
print( df["Workout_Duration_Min"].isnull().sum())
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
```

```python
df.dropna(subset=["Date"], inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.sort_values(by="Date", inplace=True)
```

```python
print(df["Date"].head())
```

```python
print( df["Date"].isnull().sum())
```

```python
df["Heart_Rate"] = df["Heart_Rate"].astype(float)
```

```python
df["Heart_Rate"].fillna(df["Heart_Rate"].median(), inplace=True)
```

```python
df.loc[(df["Heart_Rate"] < 40) | (df["Heart_Rate"] > 220), "Heart_Rate"] = df["Heart_Rate"].median()
```

```python
df["Heart_Rate"] = df["Heart_Rate"].astype(int)
```

```python
print(df["Heart_Rate"].describe())
```

```python
print( df["Heart_Rate"].isnull().sum())
```

```python
df["Sleep_Hours"] = df["Sleep_Hours"].astype(float)
```

```python
df["Sleep_Hours"].fillna(df["Sleep_Hours"].median(), inplace=True)
```

```python
df.loc[(df["Sleep_Hours"] < 0) | (df["Sleep_Hours"] > 24), "Sleep_Hours"] = df["Sleep_Hours"].median()
```

```python
df["Sleep_Hours"] = df["Sleep_Hours"].round(1)
```

```python
print(df["Sleep_Hours"].describe())
```

```python
print( df["Sleep_Hours"].isnull().sum())
```

```python
cities = ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"]
```

```python
df["City"] = [cities[i % len(cities)] for i in range(len(df))]
```

```python
df.drop_duplicates(inplace=True)
```
