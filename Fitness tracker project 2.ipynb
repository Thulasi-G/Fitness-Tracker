```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/fitness_tracker.csv")
df
```

```python
df.isnull().sum()
```

```python
df["User_ID"].fillna("Unknown", inplace=True)
user_ids = df["User_ID"].to_numpy()
unique_ids, counts = np.unique(user_ids, return_counts=True)
df["User_ID"].isnull().sum()
df["User_ID"]
```

```python
df["Name"].fillna("Unknown", inplace=True)
names = df["Name"].to_numpy()
unique_names, counts = np.unique(names, return_counts=True)
df["Name"].isnull().sum()
```

```python
median_age = np.median(df["Age"].dropna())
df["Age"].fillna(median_age, inplace=True)
ages = df["Age"].to_numpy()
unique_ages, counts = np.unique(ages, return_counts=True)
df["Age"].isnull().sum()
```

```python
df["Gender"].fillna("Unknown", inplace=True)
genders = df["Gender"].to_numpy()
unique_genders, counts = np.unique(genders, return_counts=True)
df["Gender"].isnull().sum()
```

```python
median_steps = np.median(df["Steps"].dropna())
df["Steps"].fillna(median_steps, inplace=True)
steps = df["Steps"].to_numpy()
unique_steps, counts = np.unique(steps, return_counts=True)
df["Steps"].isnull().sum()
```

```python
median_calories = np.median(df["Calories_Burned"].dropna())
df["Calories_Burned"].fillna(median_calories, inplace=True)
calories = df["Calories_Burned"].to_numpy()
unique_calories, counts = np.unique(calories, return_counts=True)
df["Calories_Burned"].isnull().sum()
```

```python
median_duration = np.median(df["Workout_Duration_Min"].dropna())
df["Workout_Duration_Min"].fillna(median_duration, inplace=True)
worked_duration = df["Workout_Duration_Min"].to_numpy()
unique_duration, counts = np.unique(worked_duration, return_counts=True)
```

```python
df["Date"].fillna(method="ffill", inplace=True)
df["Date"].fillna(method="bfill", inplace=True)
dates = df["Date"].to_numpy()
unique_dates, counts = np.unique(dates, return_counts=True)
df["Date"].isnull().sum()
```

```python
median_heart_rate = np.median(df["Heart_Rate"].dropna())
df["Heart_Rate"].fillna(median_heart_rate, inplace=True)
heart_rate = df["Heart_Rate"].to_numpy()
unique_heart_rate, counts = np.unique(heart_rate, return_counts=True)
df["Heart_Rate"].isnull().sum()
```

```python
median_sleep = np.median(df["Sleep_Hours"].dropna())
df["Sleep_Hours"].fillna(median_sleep, inplace=True)
sleep_hours = df["Sleep_Hours"].to_numpy()
unique_sleep, counts = np.unique(sleep_hours, return_counts=True)
df["Sleep_Hours"].isnull().sum()
```

```python
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
df["City"] = "Unknown"
df["City"].fillna("Unknown", inplace=True)
cities = df["City"].to_numpy()
unique_cities, counts = np.unique(cities, return_counts=True)
df["City"].isnull().sum()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


file_path = "/content/fitness_tracker.csv"
df = pd.read_csv("/content/fitness_tracker.csv")


plt.hist(df["Steps"], bins=20, edgecolor="black")
plt.title("Distribution of Steps")
plt.xlabel("Steps")
plt.ylabel("Frequency")
plt.show()


plt.hist(df["Calories_Burned"], bins=20, edgecolor="black")
plt.title("Distribution of Calories Burned")
plt.xlabel("Calories Burned")
plt.ylabel("Frequency")
plt.show()


plt.hist(df["Age"], bins=10, edgecolor="black")
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()


plt.scatter(df["Steps"], df["Calories_Burned"])
plt.title("Steps vs Calories Burned")
plt.xlabel("Steps")
plt.ylabel("Calories Burned")
plt.show()
```

```python
df["Gender"].fillna("Unknown", inplace=True)
gender_counts = df["Gender"].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Gender Distribution")
plt.axis('equal')  # Equal aspect ratio ensures pie is a circle
plt.show()
```

```python
df["Name"].fillna("Unknown", inplace=True)
gender_counts = df["Name"].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Name Distribution")
plt.axis('equal')  # Equal aspect ratio ensures pie is a circle
plt.show()
```

```python

age_counts = df["Age"].value_counts().sort_index()

plt.figure(figsize=(10, 5))
plt.plot(age_counts.index, age_counts.values, marker='o', linestyle='-', color='b')
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Number of Entries per Age")
plt.xticks(rotation=70)
plt.grid(True)
plt.tight_layout()
plt.show()


name_counts = df["Name"].value_counts().sort_index()  # Optional: sort by name alphabetically

plt.figure(figsize=(10, 5))
plt.plot(name_counts.index, name_counts.values, marker='o', linestyle='-', color='g')
plt.xlabel("Name")
plt.ylabel("Count")
plt.title("Number of Entries per Name")
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```
